读取视频，抽帧并保存

```python
import cv2 as cv
video=cv.VideoCapture('./detect/1.mp4')
num=0
save_step=30
while True:
    ret,frame=video.read()
    if not ret:
        break
    num+=1
    if num%save_step==0:
        cv.imwrite('./detect/images/'+str(num)+'.jpg',frame)
```
